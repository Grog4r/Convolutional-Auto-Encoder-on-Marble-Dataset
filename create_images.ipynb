```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def split_and_plot_rgb_channels(image_path):
    """
    Loads an RGB image, splits it into R, G, B channels,
    and plots the original and channel images.
    
    Args:
        image_path (str): Path to the input image file.
    """
    # Load image
    img = Image.open(image_path).convert('RGB')
    img_np = np.array(img)

    # Split channels
    R = img_np[:, :, 0]
    G = img_np[:, :, 1]
    B = img_np[:, :, 2]

    # Plotting
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 4, 1)
    plt.imshow(img_np)
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(R, cmap='Reds')
    plt.title("Red Channel")
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(G, cmap='Greens')
    plt.title("Green Channel")
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(B, cmap='Blues')
    plt.title("Blue Channel")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
split_and_plot_rgb_channels("marble/train/good/_0_0_65.jpg")
split_and_plot_rgb_channels("marble/test/crack/_0_0_20210531_17292_0.jpg")
```
